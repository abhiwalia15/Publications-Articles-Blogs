# Multiple Linear Regression
## step1: Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib  inline
```

## step2 :Load dataset

```python
df=pd.read_csv("housing_prices.csv")
```

```python
df
```

```python
df=df.iloc[:,[0,1,2,4]]
```

```python
df.head()
```

## Step3:Feature matrix and Target vector

```python
x=df.iloc[:,:3].values
y=df.iloc[:,3].values
```

```python
print(x[:5])
print(y[:5])
```

## step4:Split the data into 80-20

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
print(x_train.shape)
print(x_test.shape)
print(x_train.shape)
print(x_test.shape)
```

## step5: Fit the line:Train the multiple lineAR Model

```python
from sklearn.linear_model import LinearRegression
```

```python
mlr_model= LinearRegression(fit_intercept=True)
```

```python
mlr_model.fit(x_train,y_train)
```

```python
print(mlr_model.intercept_) # (PRICE=(-4481.80028058845)+8.65903854)*AREA
print(mlr_model.coef_)#y=c+mx
```

## calculating R^2score using tain and test model

```python
print(mlr_model.score(x_train,y_train))
print(mlr_model.score(x_test,y_test))
```

## vISUALIZATIONM

```python
plt.scatter(x_train[:,0],y_train,c='red')
plt.scatter(x_test[:,0],y_test,c='blue')
plt.plot(x_train[:,0],mlr_model.predict(x_train),c='y')
```
